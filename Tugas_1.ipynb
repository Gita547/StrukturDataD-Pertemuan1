```python
import numpy as np
import matplotlib.pyplot as plt

def program_nilai():
    print("=== INPUT 10 NILAI MAHASISWA ===")
    data_nilai = []
    # Kerangka awal program
    # 1. Input 10 nilai (O(n))
    for i in range(10):
        while True:
            try:
                n = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
                if 0 <= n <= 100:
                    data_nilai.append(n)
                    break
                else:
                    print("⚠️ Nilai harus 0-100.")
            except ValueError:
                print("⚠️ Masukkan angka!")
    arr_nilai = np.array(data_nilai)
    n_max = np.max(arr_nilai)
    n_min = np.min(arr_nilai)
    rata_rata = np.mean(arr_nilai)
    lulus = arr_nilai[arr_nilai >= 60]
    jml_lulus = len(lulus)
    jml_gagal = len(arr_nilai) - jml_lulus
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    ax1.bar(['Tertinggi', 'Terendah'], [n_max, n_min], color=['#2ecc71', '#e74c3c'])
    ax2.pie([jml_lulus, jml_gagal], labels=['Lulus', 'Gagal'], autopct='%1.1f%%', startangle=90)
    plt.show()

    # Jalankan Program
if __name__ == "__main__":
    program_nilai()
```
